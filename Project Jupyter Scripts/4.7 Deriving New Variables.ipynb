## Creating price range variable (without loc)

```python
# Create subset to work with
df = ords_prods_merged[:1000000]
```

```python
def price_label(row):
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range'
    else: return 'Not enough data'
```

```python
# Applying the defined function to the dataframe
df['price_range'] = df.apply(price_label, axis = 1)
```

```python
df['price_range'].value_counts()
```

```python
# Checking the most expensive item in the dataframe
df['prices'].max()
```

## Creating price range variable with loc function on subset

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

## Create column for how busy each day of the week is - using loop function

```python
ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
result = []

for value in ords_prods_merged['orders_day_of_week']:
    if value == 0:
        result.append('Busiest day')
    elif value == 4:
        result.append('Least busy')
    else:
        result.append('Regularly busy')
        
# Result is creating an empty list to then be inserted into the dataframe
```

## Create column for 'busiest days' (plural)

```python
# Creating new list for 2 most busy days and 2 least busy days

result_2 = [] 

for value in ords_prods_merged['orders_day_of_week']:
    if value == 0 or value == 1:
        result_2.append('Busiest days')
    elif value == 4 or value == 3:
        result_2.append('Least busy days')
    else:
        result_2.append('Regularly busy')
```

```python
result_2
```

```python
# Adding list to new column of dataframe
ords_prods_merged['Busiest days'] = result_2
```

```python
# Check whether it has been added
ords_prods_merged.head()
```

```python
# Checking values of new column for accuracy
ords_prods_merged['Busiest days'].value_counts(dropna = False)
```

## Create new column for busiest hours of the day

```python
# Assess the busiest times of day
ords_prods_merged['order_time_of_day'].value_counts(dropna = False)
```

Observation:

Most orders = 10, 11, 14, 15, 13, 12, 16, 9

Average orders = 17, 8, 18, 19, 20, 7, 21, 22

Fewest orders = 23, 6, 0, 1, 5, 2, 4, 3

```python
# Create list using for loop

result_3 = []

for value in ords_prods_merged['order_time_of_day']:
    if value in [10, 11, 14, 15, 13, 12, 16, 9]:
        result_3.append('Most orders')
    elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
        result_3.append('Fewest orders')
    else:
        result_3.append('Average orders')
```

```python
result_3
```

```python
# Adding list to new column 
ords_prods_merged['Busiest period of day'] = result_3
```
