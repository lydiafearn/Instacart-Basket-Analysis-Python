### Scatterplot showing age and income

```python
scatter_age_income = sns.scatterplot(data = df_ord_prod_cust, x = 'Age', y = 'income', color = 'lightskyblue')
```

Observation: The scatterplot shows that generally income increases with age. There are no individuals below the age of 40 that are earning above 400k. Whereas after 40 there are more people earning up to 600k. In addition, you can see that above 40 there are more individuals earning within the 400k and 200k bracket. Therefore, it shows that there is a correlation between higher age and more income.

```python
# Export scatterplot
scatter_age_income.figure.savefig(os.path.join(path, 'Visualisations', 'scatter_age_income.png'))
```
